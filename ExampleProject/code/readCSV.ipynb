```python
# Example Code from Konstantinos Theofilatos (23.09.2021)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read the data by loading them in a pandas "dataframe"
df = pd.read_csv('http://theofil.web.cern.ch/theofil/files/cms_dimuons.txt')
```

```python
# print first 5 events of the dataframe
df.head()
```

```python
# define a new variable, the transverse momentum (pt) realative to the z axis for the 1st particle
df['pt1'] = (df['px1']**2 + df['py1']**2)**0.5

# print first 5 events of (modified) the dataframe
df.head()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
plt.hist(df['pt1'], bins = 50, ec="black")
ax.set_ylabel('events / bin')
ax.set_xlabel('pt1 [GeV]')
plt.rc('font', size=16)
ax.set_yscale('log')
```
